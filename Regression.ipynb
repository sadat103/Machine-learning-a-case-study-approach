```python
#Create a simple regression model of sqft_living 
```

```python
train_data , test_data = sales.random_split(.8,seed=0)
```

```python
#build the regression model
```

```python
def get_residual_sum_of_squares(model, data, outcome):
    # First get the predictions
    yhats = model.predict(data)
    # Then compute the residuals/errors
    residuals = yhats - outcome
    # Then square and add them up
    RSS = (residuals*residuals).sum()
    return(RSS)    
```

```python
rss_example_train = get_residual_sum_of_squares(sqft_model, test_data, test_data['price'])
print(rss_example_train) 
```

```python
rss_example_train = get_residual_sum_of_squares(my_features_model, test_data, test_data['price'])
print(rss_example_train) 
```
